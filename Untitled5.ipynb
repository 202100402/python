```python
#202100402 권하늘 과제4_커피, 에너지 음료, 소프트 드링크, 차, 물 분류하기

from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



targetUrl="https://raw.githubusercontent.com/gundaminpde/test1/main/test2.csv"

df = pd.read_csv(targetUrl)


normalization_df=(df-df.mean(numeric_only=True))/df.std(numeric_only=True)
x=normalization_df.iloc[:,0:3]
y=df.iloc[:,4]
print(normalization_df.head())
print(y.head())

y=pd.get_dummies(y)

model = Sequential()
model.add(Dense(36, input_dim=3, activation='sigmoid'))
model.add(Dense(33, activation='relu'))
model.add(Dense(6,activation='softmax'))
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


history = model.fit(x,y, epochs=300, batch_size=30)

y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
```
